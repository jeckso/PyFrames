\begin{align*}
  x_{1}+x_{2}+2x_{3}+3x_{4}=1 \\ 
  3x_{1}-x_{2}-x_{3}-2x_{4}=-4 \\ 
  2x_{1}-3x_{2}-x_{3}-x_{4}=-6 \\
  x_{1}+2x_{2}+3x_{3}-1x_{4}=-4
\end{align*}

```python
import numpy as np
    

def solve(a, b):
    b = np.copy(b)
    det = np.linalg.det(a)
    dets = list()

    for i in range(len(a)):
        curr_matrix = np.copy(a)
        curr_matrix[:, i] = b
        current_det = np.linalg.det(curr_matrix)
        dets.append(current_det)

    return np.array(dets) / det

A = np.array([[1, 1, 2, 3],
              [3, -1, -1, -2],
              [2, -3, -1, -1],
              [1, 2, 3, -1]])

B = np.array([1, -4, -6, -4])

X1 = np.linalg.inv(A).dot(B)
X2 = np.linalg.solve(A,B)
X3 = solve(A, B)
print(X1)
print(X2)
print(X3)
print(np.allclose(X1, X2))
print(np.allclose(X1, X3))
print(np.allclose(X2, X3))
```

B
$$\begin{bmatrix} -1 & 0 & 5 \\ 0 & 1 & 3 \\ 2 & -2 & 4 \end{bmatrix}$$

```python
A = np.array([[2, 3, -1],
              [4, 5, 2],
              [-1, 0, 7]])

B = np.array([[-1, 0, 5],
              [0, 1, 3],
              [2, -2, 4]])

expression = 2*(A+B).dot(2*B-A)
X = expression

print(X)
print(expression)
print(np.allclose(X, expression))
```
