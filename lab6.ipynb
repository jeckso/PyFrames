```python
filename = 'books.csv'
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(filename)
df = df.fillna('-')
df
```

```python
def count_words(string):
    return len(string.split(' '))

df['WordsInTitle'] = df['Title'].apply(count_words)
df
```

```python
max_len = df['Title'].str.len().max()
authors = df[df['Title'].str.len() == max_len]['Author'].str.split(',').str[-1]
print(authors.to_string(index=False))
```

```python
concat_df = df['Author'].str.split(',', expand= True)[1]
regex = r'[b-df-hj-np-tv-z]{3}'
concat_df = concat_df.replace(np.NaN, '-')
filt = concat_df.str.contains(regex)
df[filt]
```

```python
filt = df['Author'].str.startswith('A')
df[filt]
```

```python
filt = df['Title'].str.lower().str.contains(r'[0-9]')
new_data = df[filt]['Title']
new_df = pd.DataFrame(new_data)
new_df
```

```python
filt = df['Title'].str.split().str[1].replace(np.NaN, '-').str.startswith('C')
df[filt]
```

```python
df_new = df['Author'].str.split(',', expand= True)[1]
regex = r'\b\w*(ea|\w(?=ea))\w*\b'
df_new = df_new.replace(np.NaN, '-')
filt = df_new.str.contains(regex)
df_new[filt]
```
